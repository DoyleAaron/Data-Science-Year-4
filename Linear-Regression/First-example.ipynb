```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# I ran into a couple of issues here with importing due to pip versions and the original seaborn style being deprecated so I imported it normally instead
# This file is just to import the necessary libraries for this example
```

```python
df = pd.read_csv('student_scores.csv')
df.head()

# The function of this is just to get pandas to read the csv correctly and print the first 5 results as a tester
```

```python
df.plot.scatter(x = 'Hours', y = 'Scores')

# This is the first initial plot of the csv data using a scatter plot
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

x = df[['Hours']]
y = df['Scores'] 

model.fit(x,y)
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)

# This uses Sci-Kit learns linear regression model to calculate the slope and fit a line to this based off of the result
# I ran into an issue on this figuring out that x needed to be a 2d array but I found the answer here (https://stackoverflow.com/questions/45554008/error-in-python-script-expected-2d-array-got-1d-array-instead)
# This now makes sense as it need to be an array of values which are encapsulated in an overall array to process them
```
